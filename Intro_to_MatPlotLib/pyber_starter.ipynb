```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
combine_data = ride_data.merge(city_data)
# Display the data table for preview
combine_data["driver_count2"] = combine_data["driver_count"]

combine_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
cities_gb = combine_data.groupby(["type","city"]).agg({"driver_count":"count","fare":"mean","driver_count2":"mean"})
cities_df = pd.DataFrame(cities_gb)
cities_df = cities_df.reset_index()
cities_df.head(20)
```

```python
# Obtain the x and y coordinates for each of the three city types
multiplyer = 12
urban_rides = []
rural_rides = []
suburban_rides = []
urban_fares = []
rural_fares = []
suburban_fares = []
urban_size = []
rural_size = []
suburban_size = []

for index, row in cities_df.iterrows():
    
    x = row["type"]
    y = row["driver_count"]
    z = row["fare"]
    b = row["driver_count2"]
    
    if x == "Urban":
        urban_rides.append(y)
        urban_fares.append(z)
        urban_size.append(b*multiplyer)
    elif x == "Rural":
        rural_rides.append(y)
        rural_fares.append(z)
        rural_size.append(b*multiplyer)
    elif x == "Suburban":
        suburban_rides.append(y)
        suburban_fares.append(z)
        suburban_size.append(b*multiplyer)
```

```python
# Build the scatter plots for each city types


urban_plot = plt.scatter(urban_rides,urban_fares,s=urban_size, color="lightcoral", label = "Urban",alpha=0.5, edgecolors="black")
rural_plot = plt.scatter(rural_rides,rural_fares,s=rural_size,color="gold",label = "Rural",alpha=0.5, edgecolors="black")
suburban_plot = plt.scatter(suburban_rides,suburban_fares,s=suburban_size,color="lightskyblue",label="Suburban",alpha=0.5,edgecolors="black")


#incorporate the other graph properties
plt.grid()

#Create a legend
legend = plt.legend(handles=[urban_plot,rural_plot,suburban_plot])
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]
# Incorporate a text label regarding circle size
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.savefig("images/scatter")
plt.show()

# Save Figure

```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_percent = combine_data.loc[combine_data["type"] == "Urban"]["fare"].sum() / combine_data["fare"].sum()
suburban_percent = combine_data.loc[combine_data["type"] == "Suburban"]["fare"].sum() / combine_data["fare"].sum()
rural_percent = combine_data.loc[combine_data["type"] == "Rural"]["fare"].sum() / combine_data["fare"].sum()

# Build Pie Chart
types_pie_chart = plt.pie([urban_percent,suburban_percent,rural_percent],labels = ["Urban","Suburban","Rural"],explode=[0.05,0,0],shadow = True,autopct="%1.1f%%",colors = ["lightcoral","lightskyblue","gold"])
plt.title("% of Total Fares by City Type")
plt.savefig("images/fare_pie")
plt.show()

# Save Figure

```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_ride_percent = combine_data.loc[combine_data["type"] == "Urban"]["driver_count"].count() / combine_data["driver_count"].count()
suburban_ride_percent = combine_data.loc[combine_data["type"] == "Suburban"]["driver_count"].count() / combine_data["driver_count"].count()
rural_ride_percent = combine_data.loc[combine_data["type"] == "Rural"]["driver_count"].count() / combine_data["driver_count"].count()
# Build Pie Chart
types_pie_chart = plt.pie([urban_ride_percent,suburban_ride_percent,rural_ride_percent],labels = ["Urban","Suburban","Rural"],explode=[0.05,0,0],shadow = True,autopct="%1.1f%%",colors = ["lightcoral","lightskyblue","gold"])
plt.title("% of Total Rides by City Type")
plt.savefig("images/rides_pie")
plt.show()
# Save Figure
```
