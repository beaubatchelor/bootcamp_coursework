```python
import tweepy
import json
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from config import consumer_key, consumer_secret, access_token, access_token_secret

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())

analyzer = SentimentIntensityAnalyzer()
```

```python
news_providers = ['BBC', 'CBS','CNN','FoxNews','nytimes']
tweet_data_list=[]
oldest_tweet = None
for news in news_providers:
    counter = 1
    
    for x in range(6):
        tweets = api.user_timeline(news, page=x)
        
        
        for tweet in tweets:
            tweet_data = {}
            tweet_data['User'] = tweet['user']['name']
            tweet_data['Tweet_Text'] = tweet['text']
            tweet_data['Created_At'] = tweet['created_at']
            tweet_data['Tweets_Ago'] = counter
            counter += 1
            oldest_tweet = tweet['id'] - 1
            tweet_data_list.append(tweet_data)
```

```python
sentiment_df = pd.DataFrame(tweet_data_list)
sentiment_df = sentiment_df[['User','Tweet_Text','Created_At','Tweets_Ago']]
sentiment_df['Compound'] = 0.0
sentiment_df['Positive'] = 0.0
sentiment_df['Neutral'] = 0.0
sentiment_df['Negative'] = 0.0
```

```python
for index, row in sentiment_df.iterrows():
    text = row[1]
    scores = analyzer.polarity_scores(text)
    sentiment_df['Compound'][index] = scores['compound']
    sentiment_df['Positive'][index] = scores['pos']
    sentiment_df['Neutral'][index] = scores['neu']
    sentiment_df['Negative'][index] = scores['neg']
    
    
    
```

```python
now = f'{datetime.now().date().month}/{datetime.now().date().day}/{datetime.now().date().year}'
sentiment_df.dtypes
```

```python
bbc_df= sentiment_df.loc[sentiment_df['User'] == 'BBC']
cbs_df= sentiment_df.loc[sentiment_df['User'] == 'CBS']
cnn_df= sentiment_df.loc[sentiment_df['User'] == 'CNN']
fox_df= sentiment_df.loc[sentiment_df['User'] == 'Fox News']
nyt_df= sentiment_df.loc[sentiment_df['User'] == 'The New York Times']
```

```python
plt.figure(figsize=(10, 8))
bbc_scatter = plt.scatter(bbc_df['Tweets_Ago'],bbc_df['Compound'],marker = "o",edgecolor = 'black',s=200,alpha=0.75)
cbs_scatter = plt.scatter(cbs_df['Tweets_Ago'],cbs_df['Compound'],marker = "o",edgecolor = 'black',s=200,alpha=0.75)
cnn_scatter = plt.scatter(cnn_df['Tweets_Ago'],cnn_df['Compound'],marker = "o",edgecolor = 'black',s=200,alpha=0.75)
fox_scatter = plt.scatter(fox_df['Tweets_Ago'],fox_df['Compound'],marker = "o",edgecolor = 'black',s=200,alpha=0.75)
nyt_scatter = plt.scatter(nyt_df['Tweets_Ago'],nyt_df['Compound'],marker = "o",edgecolor = 'black',s=200,alpha=0.75)
legend = plt.legend(handles=[bbc_scatter,cbs_scatter,cnn_scatter,fox_scatter,nyt_scatter],labels=news_providers,loc = 'right')
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]
legend.legendHandles[3]._sizes = [50]
legend.legendHandles[4]._sizes = [50]
plt.xlabel('Tweets Ago')
plt.ylabel('Tweet Polarity')
plt.title(f"Sentiment Analysis of Media Tweets ({now})")
plt.grid()
plt.savefig("images/scatter")
plt.show()
```

```python
sentiment_gb = sentiment_df.groupby('User').agg({'Compound' : 'mean'})
sentiment_gb = sentiment_gb.reset_index()
sentiment_gb.head()
```

```python
bar_chart = sentiment_gb.plot.bar(x='User',y='Compound',legend=False)
plt.ylabel('Tweet Polarity')
plt.xlabel('')
plt.title(f"Overall Media Sentiment based on Twitter ({now})")
plt.grid()
plt.savefig("images/bar")
plt.show()
```
